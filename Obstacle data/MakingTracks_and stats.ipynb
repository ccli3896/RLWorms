# permutation tests

```python
def perm_test(results, n_expts, n_successes, trials):

  wins = 0
  for i in range(trials):
    rs = np.random.permutation(results)
    if np.sum(rs[:n_expts]) >= n_successes:
      wins += 1
  
  return wins/trials
```

```python
# obstacles
results = np.concatenate([np.ones(2), np.zeros(8), np.ones(0), np.zeros(10)])
perm_test(results, 10, 6, 1000000)
```
