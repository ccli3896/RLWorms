```python
all_rewards = []
for i in np.arange(8,24,1):
     all_rewards.append(np.load('Data/Rewards_917_'+str(i)+'.json',allow_pickle=True))
all_rewards = np.hstack(all_rewards)
av_over = 1000
running = np.convolve(all_rewards, np.ones((av_over,))/av_over, mode='valid')
plt.figure(figsize=(16,5))
plt.plot(all_rewards)

plt.grid()
```

```python
plt.figure(figsize=(16,5))
plt.plot(running)
plt.title('Smoothed rewards')
plt.xlabel('Timesteps, sort of')
plt.ylabel('Pixels moved per timestep (reward)')
plt.grid();
```
